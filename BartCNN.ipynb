```python
%%writefile bart.py
import streamlit as st
from transformers import pipeline
import hashlib

# Function to hash passwords for basic security
def hash_password(password):
    return hashlib.sha256(password.encode()).hexdigest()

# User registration form
def register_user():
    st.header("User Registration")
    username = st.text_input("Username")
    password = st.text_input("Password", type="password")

    if st.button("Register"):
        # For simplicity, store user info in a dictionary (not secure for real-world use)
        users = st.session_state.setdefault("users", {})

        if username in users:
            st.warning("Username already exists. Please choose a different one.")
        else:
            hashed_password = hash_password(password)
            users[username] = {"password": hashed_password}
            st.success("Registration successful. You can now log in.")

# User login form
def login_user():
    st.header("User Login")
    username = st.text_input("Username", key="login_username")
    password = st.text_input("Password", type="password", key="login_password")

    if st.button("Login"):
        # For simplicity, check user info in the dictionary (not secure for real-world use)
        users = st.session_state.get("users", {})

        if username not in users:
            st.warning("User not found. Please register.")
        else:
            hashed_password = hash_password(password)
            if hashed_password == users[username]["password"]:
                st.success("Login successful!")
                st.session_state.logged_in = True
            else:
                st.warning("Incorrect password. Please try again.")

# Chatbot using Hugging Face's transformers
def chatbot():
    st.header("ChatBot")
    user_input = st.text_input("You: ", "")

    if st.button("Send"):
        if hasattr(st.session_state, "logged_in") and st.session_state.logged_in:
            # Get the generated response from the chat model
            response = chat_model(user_input)[0]

            # Print the entire response for inspection
            st.write("Full Response:", response)

            # Check if the key 'generated_message' exists in the response
            if 'generated_message' in response:
                st.text("Bot: " + response['generated_message'])
            else:
                st.warning("Generated message not found in the response.")
        else:
            st.warning("Please log in first.")


# Main App
def main():
    st.title("ChatBot App with Hugging Face and Streamlit")

    if not hasattr(st.session_state, "logged_in") or not st.session_state.logged_in:
        register_user()
        login_user()
    else:
        chatbot()

if __name__ == "__main__":
    # Load the chatbot model from Hugging Face
    chat_model = pipeline("text2text-generation", model="facebook/bart-large-cnn")

    # Run the Streamlit app
    st.set_page_config(page_title="ChatBot App", page_icon="🤖")
    main()
```
